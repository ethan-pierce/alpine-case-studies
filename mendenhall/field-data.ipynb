```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ssc = pd.read_csv('./data/raw/ssc.csv', delimiter = ',')
```

```python
ssc
```

```python
ssc['Site'] = [int(str(i)[0]) for i in ssc['Sample number']]
ssc['Sample'] = [int(str(i)[2]) for i in ssc['Sample number']]

ssc
```

```python
plt.style.use('default')
```

```python
plt.rcParams['font.size'] = 12
```

```python
for i in range(5):
    j = i + 1
    fig, ax = plt.subplots(1, 2, figsize = (4, 8))
    
    for idx, row in ssc[ssc['Site'] == j].iterrows():
        if row.Sample in [1.0, 2.0, 3.0, 4.0]:
            ax[0].scatter(row['Sediment mass %'] * 100, row['z'], color = 'blue', s = 80)
        else:
            ax[1].scatter(row['Sediment mass %'] * 100, row['z'], color = 'blue', s = 80)
            
    ax[0].set_title('Column ' + str(j) + 'a')
    ax[1].set_title('Column ' + str(j) + 'b')
    
    ax[0].set_xlabel('SSC (mass %)')
    ax[1].set_xlabel('SSC (mass %)')
    ax[0].set_ylabel('Height (m)')
    ax[1].set_ylabel('Height (m)')
    
    plt.tight_layout()
    plt.show()
```

```python
plt.style.use('seaborn-ticks')
plt.rcParams.update({'font.size':18})

fig, axes = plt.subplots(1, 5, figsize = (18, 8), facecolor = 'white')

for i in range(5):
    j = i + 1
    
    ax = axes[j - 1]
    
    sscvals = []
    zvals = []
    
    for idx, row in ssc[ssc['Site'] == j].iterrows():
        ax.scatter(row['Sediment mass %'] * 100, row['z'], color = 'blue', s = 80)
        
        sscvals.append(row['Sediment mass %'] * 100)
        zvals.append(row['z'])
        
    sscvals = np.array(sscvals)
    zvals = np.array(zvals)
    
    ax.plot(sscvals[zvals.argsort()], zvals[zvals.argsort()], color = 'blue', linewidth = 2, linestyle = ':')
    ax.set_facecolor('white')
        
    # ax.set_title('Column ' + str(j))

    ax.set_xlabel('SSC (mass %)')
    ax.set_ylabel('Height (m)')

plt.subplots_adjust(left = 0.1, right = 0.9, top = 0.9, bottom = 0.1, wspace = 0.4)
plt.tight_layout()
plt.show()
```
